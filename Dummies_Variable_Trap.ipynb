```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('C:\\Users\\admin\\Documents\\Housing.csv')
```

```python
df = pd.DataFrame(data=data)
df.head()
```

```python
df['furnishingstatus'].unique()
```

```python
dummies = pd.get_dummies(df.furnishingstatus)
dummies
```

```python
merged = pd.concat([df,dummies],axis=1)
merged
```

```python
final = merged.drop(['furnishingstatus'],axis = 1)
final
```

```python
final = final.drop(['unfurnished'],axis = 1)
final
```

```python
final.columns
```

```python
final['mainroad'] = final['mainroad'].map({'yes': 1, 'no': 0})
final['guestroom'] = final['guestroom'].map({'yes': 1, 'no': 0})
final['basement'] = final['basement'].map({'yes':1,'no':0})
final['hotwaterheating'] = final['hotwaterheating'].map({'yes':1,'no':0})
final['airconditioning'] = final['airconditioning'].map({'yes':1,'no':0})
final['prefarea'] = final['prefarea'].map({'yes':1,'no':0})
```

```python
final
```

```python
x = final.drop('price',axis = 1)
y = final['price']
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.predict(x)
```

```python
model.score(x,y)
```
